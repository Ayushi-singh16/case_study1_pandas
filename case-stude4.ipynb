```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "customer_id": [101, 102, 103, None, 105, 106, 107, 108],
    "name": ["Ayushi", "amit ", None, "Ravi", "Neha", "ANJALI", "rahul", None],
    "age": ["23", "twenty", "29", None, "31", "", "27", "40"],
    "city": ["Delhi", "delhi", None, "Mumbai", "Chennai", "MUMBAI", None, "Delhi"],
    "salary": ["50000", "60k", None, "70000rs", "80000", "", "90000", "1,00,000"],
    "experience": ["2 years", "3", None, "five", "6 years", "", "10", None],
    "join_date": ["2022-01-01", "01/02/2022", None, "March 5 2022", "2022/06/01", None, "2022-08-01", "invalid"],
    "is_active": ["Yes", "no", "YES", None, "No", "yes", None, "NO"]
})
```

# Identify the total number of missing values in each column.

```python
df.isnull().sum()
```

# remove rows where cutomer_id is missing 

```python
df.dropna(subset=['customer_id'])
```

# Clean the name column

```python
df['name']=df['name'].str.strip().str.lower()
```

```python
 df['name']=df['name'].fillna('unknown')
```

# Convert the age column to numeric:

```python
df['age']=pd.to_numeric(df['age'],errors='coerce')
```

```python
df['age']=df['age'].fillna(df['age'].median())
```

# Standardize the city column

```python
df['city']=df['city'].str.lower()
```

```python
df['city']=df['city'].fillna('unknown')
```

# Clean the salary column:

```python
df['salary']=df['salary'].replace("rs","",regex=False)
```

```python
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')
```

```python
df['salary'].fillna(df['salary'].mean(), inplace=True)
```

# Clean experience

```python
df['experience']=df['experience'].str.replace("years","",regex=False)
```

```python
df['experience'] = pd.to_numeric(df['experience'], errors='coerce')
```

```python
df['experience'].fillna(0, inplace=True)
```

# clean date 

```python
df['join_date'] = pd.to_datetime(df['join_date'], errors='coerce')
```

```python
df = df[df['join_date'].notna()]
```

# Clean is_active

```python
df['is_active'] = df['is_active'].str.lower()
```

```python

df['is_active'] = df['is_active'].replace({
    "yes": 1,
    "no": 0
})
```

```python
df['is_active'].fillna(0, inplace=True)
```

```python
df['is_active'] = df['is_active'].astype(int)
```

# Remove duplicates

```python
df.drop_duplicates(inplace=True)
```

# Fill numeric columns with median

```python
df = df.copy()

num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

# Fill categorical columns

```python
cat_cols = df.select_dtypes(include='object').columns
df[cat_cols] = df[cat_cols].fillna("missing")
```

# Create salary_per_year

```python
df['salary_per_year'] = df['salary'] * 12
```

# Create age_group

```python
def age_group(age):
    if age < 25:
        return "Young"
    elif age <= 35:
        return "Adult"
    else:
        return "Senior"

df['age_group'] = df['age'].apply(age_group)
```

# sort dataset

```python
df = df.sort_values(by=['salary', 'age'], ascending=[False, True])
```
